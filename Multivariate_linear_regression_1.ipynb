```python
# Remember that the equation of a linear regression multivariate have the form
# y = mx1 + mx2 + mx3 + b with each X as independent variable or feature
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprice.csv")
df
```

```python
# Handling the NAN value by computing the media
import math
median_bedrooms= math.floor(df.bedrooms.median())
median_bedrooms
```

```python
df.bedrooms= df.bedrooms.fillna(median_bedrooms)
df
```

Now we have handled our NAN values, let's create the model

```python
reg = linear_model.LinearRegression() #linear regression class object
reg.fit(df[['area', 'bedrooms', 'age']], df.price) #to train the model using the trainning set
```

```python
reg.coef_ #m values in the equation
```

```python
reg.intercept_
```

```python
# The prediction needs three inputs parameters because we have 3 X variables
reg.predict([[3000, 3, 40]])
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('hiring.csv')
df
```

```python
df.experience= df.experience.fillna("zero")
df
```

```python
!pip install word2number
from word2number import w2n # Import the word2number library
df.experience = df.experience.apply(w2n.word_to_num)
df
```

```python

import math
median_test_score = math.floor(df['test_score(out of 10)'].mean())
median_test_score
```

```python
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(median_test_score)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']],df['salary($)'])
```

```python
reg.predict([[2, 9, 6]])
```

```python
reg.predict([[10, 10, 10]])
```
